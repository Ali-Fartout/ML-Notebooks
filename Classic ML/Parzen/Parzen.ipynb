# 7-1

```python
import numpy as np 
N = 1000 
np.random.seed(1)
X = np.concatenate((np.random.normal(0, 1, int(0.3 * N)),
                    np.random.normal(5, 1, int(0.7 * N))))[:, np.newaxis]
```

# 7-2

```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian_kernel(x, x_i, h):
    return (1 / (np.sqrt(2 * np.pi) * h)) * np.exp(-0.5 * ((x - x_i) / h) ** 2)


h = 0.5 
x_values = np.linspace(-3, 8, 1000)
pdf_values = np.zeros_like(x_values)
for i, x in enumerate(x_values):
    pdf = 0
    for x_i in X:
        pdf += gaussian_kernel(x, x_i, h)
    pdf_values[i] = pdf / (N * h)  


plt.plot(x_values, pdf_values)
plt.title('Estimated Probability Density Function')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.show()
```

# 7-3

```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian_kernel(x, x_i, h):
    return (1 / (np.sqrt(2 * np.pi) * h)) * np.exp(-0.5 * ((x - x_i) / h) ** 2)



x_values = np.linspace(-3, 8, 1000)
bandwidths = [10, 1, 0.1]
plt.figure(figsize=(15, 5))

for i, h in enumerate(bandwidths):
    pdf_values = np.zeros_like(x_values)
    for j, x in enumerate(x_values):
        pdf = 0
        for x_i in X:
            pdf += gaussian_kernel(x, x_i, h)
        pdf_values[j] = pdf / (N * h)
    
    plt.subplot(1, 3, i+1)
    plt.plot(x_values, pdf_values)
    plt.title(f'Bandwidth = {h}')
    plt.xlabel('x')
    plt.ylabel('Probability Density')

plt.tight_layout()
plt.show()

```
