```python
import numpy as np
from collections import Counter
```

# 6-1

```python
class KNN:
    def __init__(self,k):
        self.k = k
        
    def fit(self,x, y):
        self.features = x
        self.targets = y
    def predict(self,x_p):
        pred = []
        for x in x_p:
            maping_features = {}
            for index, feature in enumerate(self.features):
                dist = np.linalg.norm(x - feature)
                maping_features[f"{index}-{self.targets[index]}"] = dist
            sorted_maping_features = dict(sorted(maping_features.items(), key=lambda item: item[1]))
            k_selected_keys = list(sorted_maping_features.keys())[:self.k]
            k_selected_targets = [int(key.split('-')[1]) for key in k_selected_keys]
            vote_counts = Counter(k_selected_targets)
            pred.append(max(vote_counts))
        return  np.array(pred)

        
            
        
```

# 6-3

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
data = iris.data
target = iris.target
feature_names = iris.feature_names
for i in range(len(feature_names)):
    for j in range(i + 1, len(feature_names)):
        plt.figure(figsize=(8, 6))
        plt.scatter(data[:, i], data[:, j], c=target, cmap=plt.cm.get_cmap('viridis', 3))
        plt.xlabel(feature_names[i])
        plt.ylabel(feature_names[j])
        plt.title(f'Scatter Plot: {feature_names[i]} vs {feature_names[j]}')
        plt.colorbar(label='Target Class')
        plt.show()

```

# 6-4

```python
X_train, X_valid, y_train, y_valid = train_test_split(features, target, test_size=0.4, random_state=20)
```

# 6-5

```python
knn = KNN(5)
```

```python
knn.fit(X_train, y_train)
train_pred = knn.predict(X_train)
```

```python
knn.fit(X_train, y_train)
valid_pred = knn.predict(X_valid)
```

```python
from sklearn.metrics import accuracy_score

train_accuracy = accuracy_score(y_train, train_pred)
valid_accuracy = accuracy_score(y_valid, valid_pred)
print(f"Accuracy on valid data: {valid_accuracy * 100:.2f}%")
print(f"Accuracy on train data: {train_accuracy * 100:.2f}%")
```

# 6-6

```python
from sklearn.metrics import accuracy_score


resutls = []
for k in range(1,11):
    knn = KNN(k)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_valid)
    accuracy = accuracy_score(y_valid, pred)
    resutls.append(accuracy)
```

```python
import matplotlib.pyplot as plt

plt.plot(resutls)
plt.xlabel("K")
plt.ylabel("Accuracy %")
plt.show()
```
