# Implement NaiveBayes

```python
import pandas as pd
import numpy as np
from scipy.stats import norm

class NaiveBayesClassifier:
    def __init__(self):
        self.class_probs = {}
        self.mean = {}
        self.variance = {}

    def fit(self, X, y):
        unique_classes = np.unique(y)

        for class_label in unique_classes:
            X_class = X[y == class_label]
            self.class_probs[class_label] = len(X_class) / len(X)
            self.mean[class_label] = np.mean(X_class.astype(np.float64), axis=0)
            self.variance[class_label] = np.var(X_class.astype(np.float64), axis=0)

    def calculate_likelihood(self, x, mean, variance):
        return np.prod(norm.pdf(x.astype(np.float64), loc=mean, scale=np.sqrt(variance)))

    def predict(self, X):
        likelihood = 1 
        predictions = []
        for x in X:
            x_numeric = np.array(x).astype(np.float64)  # Convert to numeric
            class_probs = []
            for class_label in self.class_probs:
                prior = self.class_probs[class_label]
                likelihood = self.calculate_likelihood(x_numeric, self.mean[class_label], self.variance[class_label])
                posterior = prior * likelihood
                class_probs.append((class_label, posterior))

            predicted_class = max(class_probs, key=lambda x: x[1])[0]
            predictions.append(predicted_class)
        return predictions
```

# Run model and validating

```python
train, test = df.iloc[:int(len(df) * 0.75) ,:], df.iloc[int(len(df) * 0.75) : ,:]
x_train, y_train = train.iloc[:,1:], train.iloc[:,0]
x_test, y_test = test.iloc[:,1:], test.iloc[:,0]
```

```python
nb_classifier = NaiveBayesClassifier()
nb_classifier.fit(x_train.values, y_train.values)
predictions = nb_classifier.predict(x_test.values)
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, confusion_matrix

accuracy = accuracy_score(y_test, predictions)
recall = recall_score(y_test, predictions, average='weighted')
precision = precision_score(y_test, predictions, average='weighted')
confusion_mat = confusion_matrix(y_test, predictions)

print("Accuracy:", accuracy)
print("Recall:", recall)
print("Precision:", precision)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
def plot_confusion_matrix(confusion_mat, classes):
    plt.figure(figsize=(8, 6))
    sns.set()
    sns.heatmap(confusion_mat, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()
```

```python
plot_confusion_matrix(confusion_mat, 2)
```
