# GMM

```python
import numpy as np
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt



def multivariate_normal_pdf(X, mean, covariance):
    n_features = X.shape[1]
    det = np.linalg.det(covariance)
    inv_covariance = np.linalg.inv(covariance)
    diff = X - mean
    exponent = -0.5 * np.einsum('...i,ij,...j->...', diff, inv_covariance, diff)
    term = (1.0 / ((2 * np.pi) ** (n_features / 2) * det ** 0.5)) * np.exp(exponent)
    return term

def expectation_step(X, means, covariances, weights):
    n_components = len(means)
    n_samples = X.shape[0]

    expectation = np.zeros((n_samples, n_components))
    for i in range(n_components):
        expectation[:, i] = weights[i] * multivariate_normal_pdf(X, means[i], covariances[i])

    expectation /= np.sum(expectation, axis=1, keepdims=True)
    return expectation

def maximization_step(X, expectation):
    n_components = expectation.shape[1]
    n_samples, n_features = X.shape
    means = np.dot(expectation.T, X) / np.sum(expectation, axis=0, keepdims=True).T

    covariances = []
    for i in range(n_components):
        diff = X - means[i]
        cov = np.dot(expectation[:, i] * diff.T, diff) / np.sum(expectation[:, i])
        covariances.append(cov)

    weights = np.mean(expectation, axis=0)

    return means, covariances, weights


```

```python
import numpy as np

def fit_gmm(X, n_components, max_iters=100, tol=1e-4, regularization=1e-6):
    n_samples, n_features = X.shape

    means = X[np.random.choice(n_samples, n_components, replace=False)]
    covariances = [np.cov(X.T) for _ in range(n_components)]
    weights = np.ones(n_components) / n_components

    for _ in range(max_iters):
        old_means, old_covariances, old_weights = means.copy(), covariances.copy(), weights.copy()

        expectations = expectation_step(X, means, covariances, weights)
        means, covariances, weights = maximization_step(X, expectations)
        covariances = [cov + regularization * np.eye(n_features) for cov in covariances]

        if np.linalg.norm(means - old_means) < tol and \
           np.allclose(covariances, old_covariances, atol=tol) and \
           np.linalg.norm(weights - old_weights) < tol:
            break

    class GMMModel:
        def __init__(self, means, covariances, weights):
            self.means = means
            self.covariances = covariances
            self.weights = weights

    return GMMModel(means, covariances, weights)
```

```python
noisy_moons, labels = make_moons(n_samples=500, noise=0.11, random_state=42)
n_components_list = [i for i in range(1,17)]
gmm_models = [fit_gmm(noisy_moons, n_components=n) for n in n_components_list]

for i, gmm_model in enumerate(gmm_models, start=1):
    print(f"==================== Component {i} ==================== ")
    print("Means: ",gmm_model.means)
    print("Covariances: ",gmm_model.covariances)
    print("Weights: ",gmm_model.weights)
```

```python
import numpy as np
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

noisy_moons, labels = make_moons(n_samples=500, noise=0.11, random_state=42)
n_components_list = [3, 8, 16]
gmm_models = [fit_gmm(noisy_moons, n_components=n) for n in n_components_list]

plt.figure(figsize=(15, 5))
for i, gmm_model in enumerate(gmm_models, start=1):
    plt.subplot(1, 3, i)
    plt.scatter(noisy_moons[:, 0], noisy_moons[:, 1], c=labels, cmap='viridis', edgecolors='k', marker='o', s=50)

    for j in range(gmm_model.means.shape[0]):
        mean = gmm_model.means[j]
        covariance = gmm_model.covariances[j]

        x, y = np.meshgrid(np.linspace(noisy_moons[:, 0].min(), noisy_moons[:, 0].max(), 100),
                           np.linspace(noisy_moons[:, 1].min(), noisy_moons[:, 1].max(), 100))
        pos = np.dstack((x, y))
        z = multivariate_normal_pdf(pos, mean, covariance)

        plt.contour(x, y, z, levels=14, colors=f'C{j}', alpha=0.5, linewidths=2)

    plt.title(f'Contour Plot (n={n_components_list[i-1]})')


plt.tight_layout()
plt.show()
```

# AIC and BIC metrics

```python
def calculate_aic_bic(X, means, covariances, weights):
    n_samples = X.shape[0]
    n_components = len(means)
    n_features = X.shape[1]

    log_likelihood = np.sum(np.log(np.sum(weights[i] * multivariate_normal_pdf(X, means[i], covariances[i])
                                          for i in range(n_components))))

    aic = -2 * log_likelihood + 2 * n_components * n_features
    bic = -2 * log_likelihood + np.log(n_samples) * n_components * n_features
    return aic, bic


def fit_gmm_with_metrics(X, n_components_range, max_iters=100, tol=1e-4, regularization=1e-6):
    best_aic = np.inf
    best_bic = np.inf
    best_model = None

    for n_components in n_components_range:
        gmm_model = fit_gmm(X, n_components, max_iters, tol, regularization)
        aic, bic = calculate_aic_bic(X, gmm_model.means, gmm_model.covariances, gmm_model.weights)

        if aic < best_aic:
            best_aic = aic
            best_model_aic = gmm_model

        if bic < best_bic:
            best_bic = bic
            best_model_bic = gmm_model

    return best_model_aic, best_model_bic


n_components_range = range(1, 16)
best_model_aic, best_model_bic = fit_gmm_with_metrics(noisy_moons, n_components_range)
```
